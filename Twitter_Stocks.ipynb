```python
short_long = {}
long_short = {}
ticker = com["ticker_symbol"]
# ticker.drop("MSFT")
name = com["company_name"]
# name.drop("Microsoft")
for i, j in zip(ticker, name):
  if i != "MSFT" and i != "TSLA":
    short_long[i] = j
    long_short[j] = i
short_long
```

```python
body = tweet["body"]
who_its_about = []
for i in body:
  curr = set()
  for j in ticker:
    if j.lower() in i.lower() and j != "MSFT" and j != "TSLA":
      curr.add(short_long[j])
  for j in name:
    if j.lower() in i.lower():
      curr.add(j)
  who_its_about += [curr]
```

```python
tweet["about_company"] = who_its_about
```

```python
freq = {}
for i in name:
  freq[i] = 0
arr = tweet["about_company"]
for i in arr:
  for j in i:  
    freq[j] += 1
freq
```

```python
dates= []
dates= tweet['date'].unique()
freq = {}
for i in dates:
  for j in name:
    freq[(i, j)] = 0
for i in dates:
  tbl = tweet[tweet["date"] == i]
  
  arr = tbl["about_company"]
  for _ in arr:
    for j in _:  
      freq[(i, j)] += 1
  
```
